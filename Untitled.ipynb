```python
train['Claim'] = train['Claim'].replace('no',0)
train['Claim'] = train['Claim'].replace('yes', 1)
```

```python
X=train.drop('Claim',axis=1)
y=train['Claim']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=7,shuffle=True,stratify=y)
```

```python
cat_feat_ind = np.where(X.dtypes==np.object)[0]
```

```python
cat_feat_ind
```

```python
import catboost
```

```python
from sklearn.metrics import*
```

```python
from catboost import CatBoostClassifier
cat = CatBoostClassifier(cat_features=cat_feat_ind,learning_rate = 0.1,iterations = 500,max_depth=10,eval_metric ='AUC',random_seed=8679)
eval_set = [(X_test,y_test)]
cat.fit(X_train,y_train,early_stopping_rounds=300,eval_set=eval_set)
y_cat=cat.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test,y_cat))
```
